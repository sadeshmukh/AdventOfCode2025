```python
# solution 2
n = 0
for id in ids:
  a, b = id.split("-")
  for i in range(int(a), int(b) + 1):
    s = str(i)
    for j in range(1, len(s)):
      if len(s) % j != 0:
        continue
      if all(s[r:r+j] == s[r+j:r+j*2] for r in range(0, len(s) - j, j)):
        n += i
        break
n
```

## Explanation
Today's was a little trickier, but nothing too bad. First, we had to check for each number in each range provided that was "doubled" (e.g. 1212), which we can find with the very convenient `s[:len(s) // 2] == s[len(s) // 2:]`. We check from the beginning to the middle (using floor division, naturally filters out odd numbers at the cost of speed).

The second half required a few more shennanigans. It wanted to check for n repetitions, e.g. including 121212 or 33333 (which are in fact odd repetitions). This was a very unoptimized solution, and it shows (taking multiple seconds to run). But it works! Using `all`  with a generator comprehension, we can enumerate every possibility, walking through the list to check if it's equal. Yes, there is a way of doing this that doesn't involve so many slices, but this was the hack that got me through.

```python
# input2
ts2 = [i.strip() for i in open("input4.txt").readlines()]
ts2[0]
l = len(ts2[0])

new = [[c == "@" for c in row] for row in ts2]
rmax = len(ts2)
cmax = len(ts2[0])

import numpy as np
from scipy.signal import convolve2d
import cv2
```

```python
%%timeit
# post advent solution, with a handful of external help

board = np.array(new, dtype=int)
filter = np.array([[1, 1, 1], [1, 0, 1], [1, 1, 1]])
t = 0

while True:
  ns = convolve2d(board, filter, mode="same", boundary="fill", fillvalue = 0)
  remove = (board == 1) & (ns < 4)
  count = np.sum(remove)
  if count == 0: break

  t += count
  board[remove] = 0
int(t)
```

```python
%%timeit
# post advent solution, with a handful of external help

board = np.array(new, dtype=int)
filter = np.array([[1, 1, 1], [1, 0, 1], [1, 1, 1]], dtype=np.uint8)
t = 0

while True:
  ns = cv2.filter2D(board, -1, filter, borderType=cv2.BORDER_CONSTANT)
  remove = (board == 1) & (ns < 4)
  count = np.sum(remove)
  if count == 0: break

  t += count
  board[remove] = 0
int(t)
```
