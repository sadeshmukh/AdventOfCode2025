```python
input = [i.strip() for i in open("input.txt", "r").readlines()]
```

```python
test_input = """L68
L30
R48
L5
R60
L55
L1
L99
R14
L82"""
input = [i.strip() for i in test_input.split("\n")]
```

```python
num = 50
e = 0
for line in input:
  if line[0] == "R":
    num += int(line[1:])
    e += num // 100
    num %= 100

  elif line[0] == "L":
    num -= int(line[1:])
    if num < 0 and num + int(line[1:]) == 0:
      num += 100
    e += abs(num // 100)
    num %= 100
    if num == 0:
      e += 1
e
```

```python
num = 50
total = 0
for line in input:
  if line[0] == "R":
    num += int(line[1:])
    total += num // 100
    num %= 100
  else:
    num -= int(line[1:])
    if num < 0 and num + int(line[1:]) == 0:
      num += 100
    while num < 0:
      num += 100
      total += 1
    if num == 0:
      total += 1
  # print(line, num, a1 if a1 else "")
total
```
