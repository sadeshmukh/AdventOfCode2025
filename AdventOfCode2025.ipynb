```python
# input
ids = [i for i in open("input2.txt").read().split(",")]
```

```python
# test input
ids = """2121212118-2121212124""".split(",")
```

```python
# solution 1
n = 0
for id in ids:
  a, b = id.split("-")
  for i in range(int(a), int(b) + 1):
    s = str(i)
    if s[:len(s) // 2] == s[len(s) // 2:]:
      n += i
n
```

```python
# solution 2
n = 0
for id in ids:
  a, b = id.split("-")
  for i in range(int(a), int(b) + 1):
    s = str(i)
    for j in range(1, len(s)):
      if len(s) % j != 0:
        continue
      if all(s[r:r+j] == s[r+j:r+j*2] for r in range(0, len(s) - j, j)):
        n += i
        break
n
```

## Explanation
Today's was a little trickier, but nothing too bad. First, we had to check for each number in each range provided that was "doubled" (e.g. 1212), which we can find with the very convenient `s[:len(s) // 2] == s[len(s) // 2:]`. We check from the beginning to the middle (using floor division, naturally filters out odd numbers at the cost of speed).

The second half required a few more shennanigans. It wanted to check for n repetitions, e.g. including 121212 or 33333 (which are in fact odd repetitions). This was a very unoptimized solution, and it shows (taking multiple seconds to run). But it works! Using `all`  with a generator comprehension, we can enumerate every possibility, walking through the list to check if it's equal. Yes, there is a way of doing this that doesn't involve so many slices, but this was the hack that got me through.

```python
# input
nums = [i.strip() for i in open("input3.txt").readlines()]
l = len(nums[0])
```

```python
# test input
test = """987654321111111
811111111111119
234234234234278
818181911112111"""

nums = [i.strip() for i in test.split("\n")]
l = len(nums[0])
nums[0]
```

```python
# part 1
e = 0

for line in nums:
  i = 0
  c = 0
  m = 0

  while i < l - 1:
    if int(line[i]) > c:
      c = int(line[i])
      m = i
    i += 1
  e += c * 10 + max(int(char) for char in line[m + 1:])
e
```

```python
# part 2
total = 0
for line in nums:
  add = []
  used = []
  need = 12
  pivot = 0
  while need > 0:
    maxn = 0
    at = 0
    for i in range(pivot, l - need + 1):
      if int(line[i]) > maxn:
        maxn = int(line[i])
        pivot = i + 1
        at = i
    add.append(maxn)
    used.append(at) # to debug when something inevitably fail
    need -= 1
  total += int("".join(str(i) for i in add))


total
```

Pretty simple. Find the biggest number (not at the end) then find the biggest number after that for the first part. I honestly small brained this by doing some shenanigans until I simplified it and it worked.
